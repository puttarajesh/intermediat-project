```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
# Load the datasets
data1 = pd.read_csv('data1.csv')
data2 = pd.read_csv('data2.csv')
data3 = pd.read_csv('data3.csv')
```

```python
# Merge data1 and data2 on the "Courier" column
merged_data = pd.merge(data1, data2, left_on='Courier', right_on='Courier', how='left')

# Merge the result with data3 on the "Weight Slab" column
merged_data = pd.merge(merged_data, data3, left_on='Weight Slab', right_on='Weight Slab', how='left')
```

```python
merged_data
```

```python
# Calculate the total shipping cost by multiplying "Charged wt" with "Multiplier" and "Additional"
merged_data['Total Cost'] = merged_data['Charged wt'] * merged_data['Multiplier'] * merged_data['Additional']
merged_data
```

```python
# Group by "Zone" and calculate the average shipping cost
average_cost_by_zone = merged_data.groupby('Zone')['Total Cost'].mean()
```
